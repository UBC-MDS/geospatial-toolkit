### Calculating Distance

The `haversine_distance` function allows you to calculate the shortest distance between two points on Earth. This is useful for shipping, flight paths, or general spatial analysis.

```python
# Define locations as (latitude, longitude) tuples
vancouver = (49.2827, -123.1207)
seattle = (47.6062, -122.3321)

# Calculate distance
dist = gst.haversine_distance(vancouver, seattle)

print(f"The distance between Vancouver and Seattle is {dist:.2f} km.")
```
